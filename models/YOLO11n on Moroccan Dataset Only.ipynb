### Hyperparameters:
The following hyperparameters are used during training:
- **Model**: `yolo11n.pt` (YOLOv11 nano model)
- **Image Size**: `640` (standard input image size)
- **Batch Size**: `64`
- **Epochs**: `100`
- **Learning Rate**:
  - `lr0 = 0.001` (initial learning rate)
  - `lrf = 0.01` (final learning rate fraction)
- **Momentum**: `0.937`
- **Weight Decay**: `0.0005`
- **Warmup Epochs**: `5`
- **Data Augmentation**:
  - **Scale**: `0.5`
  - **Degrees**: `10.0` (rotation)
  - **Translate**: `0.2`
  - **Mosaic**: `1.0`
  - **Mixup**: `0.1`

```python
data_yaml = r'/content/drive/My Drive/Moroccan_Pothole_Dataset-main/YOLO/Original/data.yaml'
images_path = r'/content/drive/My Drive/Moroccan_Pothole_Dataset-main/YOLO/Original/test/images/'
```

```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLO11n model
model = YOLO("yolo11n.pt")

results = model.train(data=data_yaml, epochs=100, imgsz=640, batch=64, augment = True)

import numpy as np
# Evaluate the model
metrics = model.val()

f1 = metrics.box.f1[0] if isinstance(metrics.box.f1, (list, np.ndarray)) else metrics.box.f1
map50 = metrics.box.map50

print(f"F1 Score: {f1}")
print(f"mAP@0.5: {map50:.4f}")

```

```python
import random
import os

# List all image files in the folder
all_images = [os.path.join(images_path, f) for f in os.listdir(images_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Randomly select multiple images (e.g., 5 images)
num_random_images = 5
random_images = random.sample(all_images, num_random_images)

# Run prediction on the selected images
results = model.predict(source=random_images, imgsz=640)

# Print or visualize results for each image
for img, res in zip(random_images, results):
    print(f"Prediction for: {img}")
    res.show()  # Visualize the prediction for the current image
```
