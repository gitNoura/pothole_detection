### Model 3 : YOLO11n Transfer Learning

### Key Features:
1. **Stage 1: Initial Training**: The model is first trained on a large dataset to learn general features.
2. **Stage 2: Fine-tuning**: The pretrained model is fine-tuned on the Moroccan dataset to adapt to local data.
3. **Evaluation**: Evaluates the model on both the large and Moroccan datasets using metrics like mAP, precision, recall, and F1 score.
4. **Device Selection**: T4-GPU on Collab.

### Hyperparameters:
The following hyperparameters are used during both stages of training:

#### Stage 1: Initial Training on Large Dataset
- **Model**: `yolo11n.pt` (YOLOv11 nano model)
- **Image Size**: `640` (standard input image size)
- **Epochs**: `50`
- **Batch Size**: `64`
- **Learning Rate**:
  - `lr0 = 0.01` (initial learning rate)
  - `lrf = 0.001` (final learning rate fraction)
- **Momentum**: `0.937`
- **Weight Decay**: `0.0005`
- **Warmup Epochs**: `3`
- **Data Augmentation**:
  - **Mosaic**: `1.0`
  - **Mixup**: `0.2`
  - **Scale**: `0.5`
  - **Degrees**: `10.0` (rotation)
  - **Translate**: `0.2`
- **Save Settings**: Save model at the end of training with the name `stage1_training`

#### Stage 2: Fine-tuning on Moroccan Dataset
- **Epochs**: `30`
- **Batch Size**: `4`
- **Learning Rate**:
  - `lr0 = 0.001` (initial learning rate)
  - `lrf = 0.0001` (final learning rate fraction)
- **Momentum**: `0.937`
- **Weight Decay**: `0.0005`
- **Warmup Epochs**: `2`
- **Data Augmentation**:
  - **Mosaic**: `0.5`
  - **Mixup**: `0.1`
  - **Scale**: `0.5`
- **Save Settings**: Save model at the end of fine-tuning with the name `stage2_finetuning`

````python
# Load the model with weights from the previous training
data_yaml1 = r'/content/drive/My Drive/Moroccan_Pothole_Dataset-main/YOLO/Original/data.yaml'
model = YOLO(r'/content/drive/My Drive/Combined_Dataset/train3/weights/best.pt')

# Fine-tune on Moroccan dataset
model.train(
    data= data_yaml1,
    epochs=50,
    imgsz=640,
    batch=16
)
import numpy as np
# Evaluate the model
metrics = model.val()

f1 = metrics.box.f1[0] if isinstance(metrics.box.f1, (list, np.ndarray)) else metrics.box.f1
map50 = metrics.box.map50

print(f"F1 Score: {f1}")
print(f"mAP@0.5: {map50:.4f}")
````

````python
import random
import os

# List all image files in the folder
all_images = [os.path.join(images_path, f) for f in os.listdir(images_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Randomly select multiple images (e.g., 5 images)
num_random_images = 5
random_images = random.sample(all_images, num_random_images)

# Run prediction on the selected images
results = model.predict(source=random_images, imgsz=640)

# Print or visualize results for each image
for img, res in zip(random_images, results):
    print(f"Prediction for: {img}")
    res.show()  # Visualize the prediction for the current image
````
