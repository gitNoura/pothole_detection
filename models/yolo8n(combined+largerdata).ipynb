### YOLOv8n
The model used in this training process is YOLOv8n, a lightweight variant of the YOLOv8 family, specifically designed for fast inference and efficient performance on edge devices.

The training process is configured as follows:
- **Data**: Custom dataset specified in the `data_yaml` file.
- **Epochs**: 100 iterations over the training dataset.
- **Image Size**: 640x640 pixels for input image resizing.
- **Batch Size**: 64 images per batch for optimized GPU utilization.
- **Augmentation**: Enabled to enhance model robustness by creating diverse training scenarios.

### Metrics Evaluated
1. **F1 Score**
2. **mAP@0.5**: The map at an (IoU) threshold of 0.5.

```python
from ultralytics import YOLO
import numpy as np
```

```python
data_yaml = r'/content/drive/My Drive/pothole_dataset_yolo/data.yaml'
images_path = r'/content/drive/My Drive/pothole_dataset_yolo/test/images/'
```

```python
# Initialize the model
model = YOLO('yolov8n.pt')  # Pre-trained YOLOv8n model

# Train the model
model.train(data=data_yaml, epochs=100, imgsz=640, batch = 64, augment= True)

metrics = model.val()
f1 = metrics.box.f1[0] if isinstance(metrics.box.f1, (list, np.ndarray)) else metrics.box.f1
map50 = metrics.box.map50

print(f"F1 Score: {f1}")
print(f"mAP@0.5: {map50:.4f}")
```

```python
import random
import os

# List all image files in the folder
all_images = [os.path.join(images_path, f) for f in os.listdir(images_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Randomly select multiple images (e.g., 5 images)
num_random_images = 5
random_images = random.sample(all_images, num_random_images)

# Run prediction on the selected images
results = model.predict(source=random_images, imgsz=640)

# Print or visualize results for each image
for img, res in zip(random_images, results):
    print(f"Prediction for: {img}")
    res.show()  # Visualize the prediction for the current image
```
